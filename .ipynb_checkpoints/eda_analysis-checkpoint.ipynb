```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("dataset/train.csv")
test = pd.read_csv("dataset/test.csv")
```

```python
loanId = test.Loan_ID
test = test.drop('Loan_ID', axis = 1)
```

```python
train = train.drop('Loan_ID', axis = 1)
```

```python
train['LoanAmount'] = train['LoanAmount'].fillna(train['LoanAmount'].mean())
```

```python
test['LoanAmount'] = test['LoanAmount'].fillna(test['LoanAmount'].mean())
```

```python
train['Loan_Amount_Term'] = train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mean())
test['Loan_Amount_Term'] = test['Loan_Amount_Term'].fillna(test['Loan_Amount_Term'].mean())
train['Credit_History'] = train['Credit_History'].fillna(method = 'ffill')
test['Credit_History'] = test['Credit_History'].fillna(method = 'ffill')
```

```python
train.head()
```

```python
train['ApplicantIncome_by_loanAmt'] = train['ApplicantIncome'] / train['LoanAmount']
train['Loan_Amount_Term_By_amt'] = train['LoanAmount'] / train['Loan_Amount_Term']
test['ApplicantIncome_by_loanAmt'] = test['ApplicantIncome'] / test['LoanAmount']
test['Loan_Amount_Term_By_amt'] = test['LoanAmount'] / test['Loan_Amount_Term']
```

```python
y = train.Loan_Status
X = train.drop('Loan_Status', axis = 1)
```

```python
X = pd.get_dummies(X)
```

```python
test = pd.get_dummies(test)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(n_estimators=800, max_features="sqrt", n_jobs=-1, max_depth=10, min_samples_split=10)
model.fit(X, y)
y_pred = model.predict(X_val)
accuracy_score(y_val, y_pred)
```

```python
import xgboost as xgb

clf = xgb.XGBClassifier(scale_pos_weight=1,
                      learning_rate=0.01,  
                      colsample_bytree = 0.9,
                      subsample = 0.9,
                      objective='binary:logistic', 
                      n_estimators=700, 
                      reg_alpha = 0.3,
                      max_depth=3, 
                      gamma=10,
                      eval_metric="logloss")
clf.fit(X, y)
pred = clf.predict(X_val)
accuracy_score(y_val, pred)
```

```python
import lightgbm
```

```python
l = lightgbm.LGBMClassifier(subsample = 0.9, max_depth=20, reg_alpha=0.5, num_leaves=60, boosting_type="gbdt")
l.fit(X, y)
y_pred1 = l.predict(X_val)
accuracy_score(y_val, y_pred1)
```

```python
test_pred = l.predict(test)
```

```python
output = pd.DataFrame({'Loan_ID': loanId,
                       'Loan_Status': test_pred})
```

```python
output.to_csv('submission.csv', index=False)
```
